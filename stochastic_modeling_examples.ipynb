```python
#1 uniform distribution between zero and one
#generate 10,000 random values 
#determine relative frequency of these random values within each class size
#recommend 100 equally spaced class sizes between -3 and 3
!pip install numpy as np
s = np.random.default_rng().uniform(0,1,10000)
np.all(s >= 0)
np.all(s < 1)

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 15, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.show()
```

```python
#graph number 1: uniformly distributed with values between zero and one
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
a = sns.distplot(random.uniform(size= 10000), 100, hist=True, label='uniform')
plt.show(a)
```

```python
#graph #4 with normally distributed with a mean of zero and variance of one
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot(random.normal(loc=0, scale=1, size=10000), 100, hist=True, label='normal') 
```

```python
#graph #2 with exponentially distributed with a mean of one
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot(random.exponential(scale=1, size=10000), 100, hist=True, label='exponential') 
```

```python
#graph #3 with extreme value type I distributed with a mean of zero and a variance of one
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot(random.gumbel(loc=0, scale=1, size=10000), 100, hist=True, label='standard Gumbel distribution')
```

```python
#graph #5 with log-normally distributed with a mean and a variance of the transformed variable of zero and one
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot(random.lognormal(0, 1, 10000), 100, hist=True, label='log normal')
```

```python
#skewed normal distribution with a mean of one, variance of one and a skew coefficient of 2
from scipy.stats import skewnorm
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(skewnorm.rvs(2, size=1000), 100, hist=True, label='skewed normal')
```
