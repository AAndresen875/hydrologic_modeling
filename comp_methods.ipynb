```python
#Chapter 1 problem 3

#Write a computer program using the bisection method or the secant method to compute the
#flow depth from Manning's equation, as discussed in class. Select three combinations of the
#parameters of flow rate, Manning's n, slope, and channel width. Compare the flow computed
#with the depths obtained with the bisection or secant methods with that specified in your
#parameter set.

#bisection Method:
def bisection(f,xp,xn,N):
    '''Approximate solution of f(x)=0 on interval [xp,xn] by bisection method.

    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    xp,xn : numbers
        The interval in which to search for a solution. The function returns
        None if f(xp)*f(xn) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    xm_N : number
        The midpoint of the Nth interval computed by the bisection method. The
        initial interval [xp_0,xn_0] is given by [xp,xn]. If f(xm_n) == 0 for some
        midpoint xm_n = (xp_n + xn_n)/2, then the function returns this solution.
        If all signs of values f(xp_n), f(xn_n) and f(xm_n) are the same at any
        iteration, the bisection method fails and return None.

    '''
    if f(xp)*f(xn) >= 0:
        print("Bisection method fails.")
        return None
    xp_n = xp
    xn_n = xn
    for n in range(1,N+1):
        xm_n = (xp_n + xn_n)/2
        f_xm_n = f(xm_n)
        if f(xp_n)*f_xm_n < 0:
            xp_n = xp_n
            xn_n = xm_n
        elif f(xn_n)*f_xm_n < 0:
            xp_n = xm_n
            xn_n = xn_n
        elif f_xm_n == 0:
            print("Found exact solution.")
            return xm_n
        else:
            print("Bisection method fails.")
            return None
    return (xp_n + xn_n)/2

f = lambda x: x**2 - x - 1
approx_phi = bisection(f,1,2,25)
print(approx_phi)
```

```python
#Chapter 1 problem 3

#Write a computer program using the bisection method or the secant method to compute the
#flow depth from Manning's equation, as discussed in class. Select three combinations of the
#parameters of flow rate, Manning's n, slope, and channel width. Compare the flow computed
#with the depths obtained with the bisection or secant methods with that specified in your
#parameter set.

#secant method
def secant(f,a,b,N):
    '''Approximate solution of f(x)=0 on interval [a,b] by the secant method.

    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    a,b : numbers
        The interval in which to search for a solution. The function returns
        None if f(a)*f(b) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    m_N : number
        The x intercept of the secant line on the the Nth interval
            m_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        The initial interval [a_0,b_0] is given by [a,b]. If f(m_n) == 0
        for some intercept m_n then the function returns this solution.
        If all signs of values f(a_n), f(b_n) and f(m_n) are the same at any
        iterations, the secant method fails and return None.
    '''
    
    if f(a)*f(b) >= 0:
        print("Secant method fails.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        f_m_n = f(m_n)
        if f(a_n)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Found exact solution.")
            return m_n
        else:
            print("Secant method fails.")
            return None
    return a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))

f = lambda x: x**2 - x - 1
secant(f,1,2,5)
```
